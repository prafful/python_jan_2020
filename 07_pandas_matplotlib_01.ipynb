```python
views = [127, 128, 254, 658, 958, 500, 999]
days = range(1,8)


#x,y
plt.plot(days, views, label='Channel Views', color='red', marker='p', markerfacecolor='b', linestyle='-.', linewidth=2)

#label
plt.xlabel('Day Number')
plt.ylabel('Views')

#legend
plt.legend()
#title
plt.title('Channel Views In A Week')



#plt.show()
```

```python
facebook_views = [127, 128, 254, 658, 958, 500, 999]
youtube_views = [225, 555, 666, 888, 777, 444, 1001]
twitter_views = [111, 222, 333, 444, 222, 555, 888]

days = range(1,8)

plt.plot(days, facebook_views, label='Facebook Views',color='r', marker='D' )
plt.plot(days, youtube_views, label='Youtube Views',color='g', marker='s')
plt.plot(days, twitter_views, label='Twitter Views', color='b', marker='o')

#label
plt.xlabel('Day Number')
plt.ylabel('Views')

#legend
plt.legend(loc="upper left")
#title
plt.title('Channel Views In A Week')

plt.show()
```
