```python
views = [127, 128, 254, 658, 958, 500, 999]
days = range(1,8)


#x,y
plt.plot(days, views, label='Channel Views', color='red', marker='p', markerfacecolor='b', linestyle='-.', linewidth=2)

#label
plt.xlabel('Day Number')
plt.ylabel('Views')

#legend
plt.legend()
#title
plt.title('Channel Views In A Week')



#plt.show()
```
